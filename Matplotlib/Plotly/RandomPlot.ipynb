```python
#program to draw a scatter plot for random 1000 x and y coordinates
```

```python
import random
import matplotlib.pyplot as plt 
import plotly.express as px
```

```python
x= [random.triangular() for i in range(1000)]
y= [random.triangular() for i in range(1000)]
```

```python
figure = px.scatter(x=x,y=y)
figure.update_layout(title="Random Distribution")
figure.show()

```
