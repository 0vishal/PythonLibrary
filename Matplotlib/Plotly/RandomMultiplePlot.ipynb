```python
#program to draw line and scatter plots for random 100 x and y coordinates
```

```python
import plotly.graph_objects as go
import random
```

```python
x = [random.triangular() for i in range(100)]
y = [random.gauss(0.5,0.25) for i in range(100)]
```

```python
figure = go.Figure()
figure.add_trace(go.Scatter(x=x, y=y,mode='markers'))
figure.add_trace(go.Line(x=x,y=y))
figure.update_layout(title='random distribution')
figure.show()
```
