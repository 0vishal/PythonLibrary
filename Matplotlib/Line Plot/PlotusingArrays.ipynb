```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
data = np.arange(0,20,2)
print(data)
```

```python
plt.plot(data,data,'b*',data,data*2,'r--')
plt.axis([0,40,0,40])
```
