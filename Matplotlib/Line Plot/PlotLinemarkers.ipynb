```python
#Python program to plot two or more lines and set the line markers.
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
x = [1,3,5]
y = [3,5,6]
x2 = [2,3,6]
y2 = [3,4,7]
plt.plot(x,y,label="First",linewidth=3,color='red',linestyle='dotted',marker='o',markerfacecolor='blue',markersize='6')
plt.plot(x2,y2,label="Second",linewidth=2,color='blue',linestyle='dashed',marker='o',markerfacecolor='red',markersize='4')
plt.ylim(1,8)
plt.xlim(1,8)
plt.legend()
plt.show()
```
