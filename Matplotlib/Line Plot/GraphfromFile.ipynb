```python
#program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
with open("test.txt") as file:
    data = file.read()
data = data.split('\n')    
```

```python
x = [row.split(' ')[0] for row in data]
y = [row.split(' ')[1] for row in data]
```

```python
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Data points from file")
```
