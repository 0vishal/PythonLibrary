```python
#Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.
```

```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
men_means = (22, 30, 33, 30, 26)
women_means = (25, 32, 30, 35, 29)
index = np.arange(len(men_means))
width = 0.3
figure, axes = plt.subplots()
rects1 = axes.bar(index - width/2, men_means, width,
                label='Men')
rects2 = axes.bar(index + width/2, women_means, width, 
                label='Women')
axes.set_ylabel('Scores')
axes.set_title('Scores by group and gender') 
axes.set_xticks(index)
axes.set_xticklabels(('G1', 'G2', 'G3', 'G4', 'G5'))
axes.legend()               
```
