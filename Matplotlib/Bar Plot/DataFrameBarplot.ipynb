```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
array = np.array([[4,8,5,7,6],[2,3,4,2,6],[4,7,4,7,8],[2,6,4,8,6],[2,4,3,3,2]])
data = pd.DataFrame(array, columns=['a','b','c','d','e'], index=[2,4,6,8,10])
print(data)
```

```python
data.plot(kind='bar')
```
