```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
language = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
```

```python
plt.bar(language,popularity,color=['red', 'blue', 'purple', 'green', 'lavender','black'])
plt.xlabel("Popularity")
plt.ylabel("Language")
```
