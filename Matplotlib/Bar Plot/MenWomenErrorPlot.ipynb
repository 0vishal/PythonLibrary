```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
menMeans = (22, 30, 35, 35, 26)
womenMeans = (25, 32, 30, 35, 29)
menStd = (4, 3, 4, 1, 5)
womenStd = (3, 5, 2, 3, 3)
index = np.arange(len(menMeans))
```

```python
figure, axes = plt.subplots()
menbar = plt.bar(index,menMeans,width=0.5, yerr=menStd, color='red')
womenbar = plt.bar(index,womenMeans,width=0.5,bottom=menMeans, yerr=womenStd, color='green')
plt.xlabel("Group")
plt.ylabel("Score")
plt.xticks(index,('Group1', 'Group2', 'Group3', 'Group4', 'Group5'))
plt.yticks(np.arange(0,100,10))
plt.title('Scores by group and gender')
plt.show()
```
