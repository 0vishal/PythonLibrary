```python
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
%matplotlib inline
```

```python
Means = (54.74, 42.35, 67.37, 58.24, 30.25)
Std = (4, 3, 4, 1, 5)
x_position = np.arange(len(Means))
```

```python
figure, axes = plt.subplots()
axes.bar(x_position,Means,yerr=Std,width=0.6)
for i in range(len(x_position)):
    plt.text(i,Means[i],Means[i],ha="center",va='bottom')
plt.ylabel('Scores')
plt.xlabel('Velocity')
```
