```python
from matplotlib import pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
x = [1,3,5]
y = [3,5,6]
x2 = [2,3,6]
y2 = [3,4,7]
plt.plot(x,y,label="First")
plt.plot(x2,y2,label="Second")
plt.legend()
plt.show()
```
