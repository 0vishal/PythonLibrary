```python
#programming to create a pie chart with a title of the popularity of programming Languages.
```

```python
import matplotlib.pyplot as plt
```

```python
languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
colors = ['red','blue','white','orange','green','yellow','purple']
explode = (0.1, 0, 0, 0, 0, 0)
```

```python
plt.pie(popularity,colors=colors,explode=explode,labels=languages)
plt.title("Popularity of Programming languages")
```
