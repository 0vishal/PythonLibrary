```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
medal_data = pd.read_csv('medal.csv')
print(medal_data)
```

```python
country = medal_data['country']
gold_medal = medal_data['gold_medal']
colors = ['red','blue','orange','green','yellow']
explode = (0.1, 0, 0, 0, 0)
```

```python
plt.pie(gold_medal,explode=explode,colors=colors,labels=country,shadow=True, startangle=140,autopct='%1.1f%%')
```
