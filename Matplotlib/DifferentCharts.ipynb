```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
```

```python
x = [1,3,5]
y = [3,5,6]
x2 = [2,3,6]
y2 = [3,4,7]
```

```python
plt.plot(x,y,x2,y2)
plt.axis([1,8,1,8])
plt.show()
```

```python
data = [1,2,3,4,5,6,7,4,2,5]
plt.hist(data)
```

```python
exceldata = pd.read_csv('fdata.csv')
print(exceldata)
```

```python
plt.bar(exceldata['Date'],exceldata['Close'])
plt.xlabel("Date")
plt.ylabel("Close")
```

```python
price = pd.Series([100, 200 , 300, 400, 200, 300, 1000])
print(price)
plt.boxplot(price)
```

```python
prices = [100, 200 , 300, 400]
pricelabel = ['100','200','300','400']
plt.pie(prices,labels=pricelabel)
```
